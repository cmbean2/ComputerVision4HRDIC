```python
import torch

# Check if CUDA is available
cuda_available = torch.cuda.is_available()

# Print the number of GPUs available
num_gpus = torch.cuda.device_count()

# Print the name of the GPU
gpu_name = torch.cuda.get_device_name(0) if cuda_available else "No GPU"

print(f"CUDA available: {cuda_available}, Number of GPUs: {num_gpus}, GPU Name: {gpu_name}")

print(torch.version.cuda)
```

```python
import torch

# Clear cache memory
torch.cuda.empty_cache()

# Reset peak memory stats
torch.cuda.reset_peak_memory_stats()

# Optionally, you can also collect garbage
import gc
gc.collect()
```

```python
import subprocess

# List running processes
processes = subprocess.check_output(['nvidia-smi', '--query-compute-apps=pid,process_name,used_memory', '--format=csv,noheader']).decode('utf-8').strip().split('\n')

# Kill non-essential processes (e.g., python.exe)
for process in processes:
    details = process.split(',')
    pid = details[0].strip()
    process_name = details[1].strip()
    
    if 'python.exe' in process_name:
        try:
            subprocess.call(['taskkill', '/PID', pid, '/F'])
            print(f'Terminated process {process_name} with PID {pid}')
        except Exception as e:
            print(f'Could not terminate process {process_name} with PID {pid}: {e}')
```
